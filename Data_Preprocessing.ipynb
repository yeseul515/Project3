```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgba
```

```python
df = pd.read_csv('/content/final.csv', encoding='cp949')
df
```

```python
df = df.set_index(df.columns[0])
df
```

```python
df['sum'] = df.sum(axis=1)
df
```

```python
# 리뷰 수에 대한 히스토그램 확인

sns.histplot(df['sum'], bins=10, color='gold')
```

```python
# 로그변환

log_df = np.log(df['sum'])
log_df
```

```python
#로그 변환 데이터 boxenplot으로 시각화

mediumseagreen_rgb = (0.235, 0.702, 0.443)
seaborn_mediumseagreen = sns.color_palette([to_rgba(mediumseagreen_rgb)])
sns.boxenplot(y=log_df, palette=seaborn_mediumseagreen)
plt.show()
```

```python
# 기준 구하는 사분위수

q_25 = log_df.quantile(0.25)
q_75 = log_df.quantile(0.75)

IQR = q_75-q_25

q1 = q_25-IQR*1.5
print('q1: ', q1)

original_q1 = np.exp(q1)
print('original_q1: ', original_q1)
```
